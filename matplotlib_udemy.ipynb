```python
hd=pd.read_csv("heart-disease.csv")
hd
```

```python
hd.head()
```

```python
hd.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50=hd[hd["age"]>50]
over_50
```

```python
len(over_50)
```

```python
over_50.plot.scatter(x="age",y="chol",c="target");
```

```python
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot.scatter(x="age",y="chol",c="target",ax=ax);
ax.set_xlim=([45,100]);
```

```python
fig, ax=plt.subplots()
ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

ax.set(title='heart disease and cholestrol level',xlabel='age',ylabel='cholestrol');
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter=ax0.scatter(over_50['age'],over_50['chol'],c=over_50['target'],cmap='winter')
scatter;
ax0.set(title="age vs chol",xlabel='age',ylabel='cholestrol');
ax0.legend(*scatter.legend_elements(),title='Target');
ax0.axhline(y=over_50['chol'].mean(),linestyle='--');

scatter=ax1.scatter(over_50['age'],over_50['thalach'],c=over_50['target'])
scatter;
ax1.set(title="age vs max heart rate",xlabel='age',ylabel='max heart rate');
ax1.legend(*scatter.legend_elements(),title='Target');
ax1.axhline(y=over_50['thalach'].mean(),linestyle='--');

fig.suptitle('heart disease analysis',fontsize=18,fontweight='bold');
```
